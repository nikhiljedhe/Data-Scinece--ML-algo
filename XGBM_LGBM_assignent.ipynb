```python
train['Age']=train['Age'].fillna(train['Age'].median())
test['Age']=test['Age'].fillna(test['Age'].median())

train['Cabin']=train['Cabin'].fillna(train['Cabin'].mode()[0])
test['Cabin']=test['Cabin'].fillna(test['Cabin'].mode()[0])

train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])

test['Fare']=test['Fare'].fillna(test['Fare'].median())
```

```python
# Feature Engineering/Preparation:
# - Categorical features ('Sex', 'Embarked', 'Cabin') are label encoded for both training and test sets.
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in ['Sex', 'Embarked', 'Cabin']:
    train[col]=le.fit_transform(train[col])
    test[col]=le.fit_transform(test[col])
#train['Sex', 'Embarked']=le.fit_transform(train['Sex', 'Embarked'])
#test['Sex', 'Embarked']=le.fit_transform(test['Sex', 'Embarked'])
```

```python
features=[  'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']
X=train[features]
y=train['Survived']
X_test=test[features]
```

```python
# Model Training and Evaluation:
# - Two models were trained: XGBoost (`xgb.XGBClassifier`) and LightGBM (`lgb.LGBMClassifier`).
# - Both models were trained on the same features: 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked'.
# - The target variable is 'Survived'.
# - The models were evaluated based on their performance on the training data using `accuracy_score` and `classification_report`.
```

```python
import xgboost as xgb
xgb_model=xgb.XGBClassifier(n_estimators=1000, learning_rate=0.05)
xgb_model.fit(X,y)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('LightGBM Precision:', precision_score(y,lgb_model.predict(X)))
print('XGBoost Precision:', precision_score(y,xgb_model.predict(X)))

print('LightGBM Recall:', recall_score(y,lgb_model.predict(X)))
print('XGBoost Recall:', recall_score(y,xgb_model.predict(X)))

print('LightGBM F1-score:', f1_score(y,lgb_model.predict(X)))
print('XGBoost F1-score:', f1_score(y,xgb_model.predict(X)))
```

```python
import matplotlib.pyplot as plt

# Define the metrics
metrics = ['Precision', 'Recall', 'F1-score']
lightgbm_scores = [0.9766, 0.9766, 0.9766]
xgboost_scores = [0.9820, 0.9561, 0.9689]

# Create a bar chart
x = range(len(metrics))
width = 0.35

plt.bar([i - width/2 for i in x], lightgbm_scores, width, label='LightGBM')
plt.bar([i + width/2 for i in x], xgboost_scores, width, label='XGBoost')

plt.xticks(x, metrics)
plt.ylabel('Score')
plt.title('Model Comparison')
plt.legend()
plt.show()
```

#Model Comparison
Both LightGBM and XGBoost models performed well, with high precision, recall, and F1-scores.

- XGBoost has slightly higher precision (0.982 vs 0.977).
- LightGBM has slightly higher recall (0.977 vs 0.956).
- LightGBM has a slightly higher F1-score (0.977 vs 0.969).

Overall, both models are strong performers, with XGBoost having a slight edge in precision and LightGBM having a slight edge in recall and F1-score.

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid for LightGBM
param_grid_lgb = {
    'num_leaves': [31, 62, 127],
    'max_depth': [-1, 5, 10],
    'learning_rate': [0.01, 0.1, 0.5]
}

# Define hyperparameter grid for XGBoost
param_grid_xgb = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.5],
    'n_estimators': [50, 100, 200]
}
```

```python
grid_search_lgb = GridSearchCV(lgb.LGBMClassifier(), param_grid_lgb, cv=5)
grid_search_xgb = GridSearchCV(xgb.XGBClassifier(), param_grid_xgb, cv=5)

grid_search_lgb.fit(X, y)
grid_search_xgb.fit(X, y)

print('Best Parameters for LightGBM:', grid_search_lgb.best_params_)
print('Best Parameters for XGBoost:', grid_search_xgb.best_params_)
```
