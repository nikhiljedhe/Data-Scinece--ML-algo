```python
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['LB'])
plt.title('Distribution of Baseline Fetal Heart Rate')
plt.show()
```

```python
sns.boxplot(df['AC'])
plt.title('Distribution of Accelerations')
plt.show()
```

```python
# Scatter plots for relationships between variables
sns.scatterplot(x='LB', y='AC', data=df)
plt.title('Relationship between Baseline Fetal Heart Rate and Accelerations')
plt.show()
```

```python
# Correlation heatmap
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#Insights and patterns
- Increasing trend in Baseline Fetal Heart Rate: This could indicate fetal distress or other complications.
- Decreasing trend in Accelerations: This could indicate fetal compromise or other issues.

- Strong positive correlation between LB (Baseline Fetal Heart Rate) and AC (Accelerations): This suggests that when the baseline fetal heart rate is high, there are more accelerations, which could indicate fetal well-being.
- Negative correlation between LB and DL (Decelerations Late): This suggests that when the baseline fetal heart rate is low, there are more late decelerations, which could indicate fetal distress.
- Positive correlation between UC (Uterine Contractions) and DL: This suggests that uterine contractions may be associated with late decelerations, which could indicate fetal distress.

1.LB - Likely stands for "Baseline Fetal Heart Rate (FHR)" which represents the average fetal heart rate over a period.

2.AC - Could represent "Accelerations" in the FHR. Accelerations are usually a sign of fetal well-being.

3.FM - May indicate "Fetal Movements" detected by the monitor.

4.UC - Likely denotes "Uterine Contractions", which can impact the FHR pattern.

5.DL - Could stand for "Decelerations Late" with respect to uterine contractions, which can be a sign of fetal distress.

6.DS - May represent "Decelerations Short" or decelerations of brief duration.

7.DP - Could indicate "Decelerations Prolonged", or long-lasting decelerations.

8.ASTV - Might refer to "Percentage of Time with Abnormal Short Term Variability" in the FHR.

9.MSTV - Likely stands for "Mean Value of Short Term Variability" in the FHR.

10.ALTV - Could represent "Percentage of Time with Abnormal Long Term Variability" in the FHR.

11.MLTV - Might indicate "Mean Value of Long Term Variability" in the FHR.

Correlation Analysis:-



     Negative Correlation between LB and AC

     Positive Correlation between DL and MSTV
With a correlation score of 0.43 between DL (Decelerations Late) and MSTV (Mean Value of Short Term Variability), there is a moderate positive correlation between the two variables.

Potential Implications
This correlation suggests that as the frequency or severity of late decelerations increases, the mean value of short-term variability in fetal heart rate also tends to increase.
