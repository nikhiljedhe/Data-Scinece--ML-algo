**Interview Questions**

1. What is Normalization & Standardization and how is it helpful?



Normalization and standardization are techniques used to scale the data to a common range. Normalization typically scales the data between 0 and 1, while standardization scales the data to have a mean of 0 and a standard deviation of 1. These techniques are helpful in preventing features with large ranges from dominating the model, improving the stability and performance of the model.

2. What techniques can be used to address multicollinearity in multiple linear regression?



Several techniques can be used to address multicollinearity, including:

- Variance Inflation Factor (VIF): Calculate the VIF for each feature to identify highly correlated features.
- Correlation analysis: Analyze the correlation matrix to identify highly correlated features.
- Feature selection: Select a subset of features that are not highly correlated.
- Regularization techniques: Use regularization techniques like Lasso or Ridge regression to reduce the impact of multicollinearity.
